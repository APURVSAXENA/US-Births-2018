```python
def get_percentages(col):
    print(round(df[col].value_counts()/len(df)*100,2))
```

```python
get_percentages('NO_RISKS')
```

```python
for i, col in enumerate(df.columns):
    if df[col].nunique() < 20:
        print('Column #',i,':',col)
        get_percentages(col)
        print('---------------------------')
    else:
        print('Column #',i,':',col)
        plt.hist(df[col])
        plt.show()
```
